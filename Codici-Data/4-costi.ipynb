```python
import matplotlib.pyplot as plt

countries = ["Germania", "Regno Unito", "Francia", "Italia"]
gdp_values = [4.07, 3.07, 2.78, 2.01] 

countries.append("Costo Totale Transizione Ecologica")
gdp_values.append(150)

plt.figure(figsize=(12, 6))
plt.bar(countries, gdp_values, color=['#5e9cff', '#5e9cff', '#5e9cff', '#5e9cff', '#5eff84'])

plt.xlabel('Nazioni / Costo Totale Tansizione Ecologica')
plt.ylabel('Trillions USD')
plt.title('PIL dei Maggiori Paesi Europei vs Costo Totale Transizione Ecologica (2024-2050)')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.show()
```

```python
years = [2025, 2030, 2035, 2040, 2045, 2050]
world_gdp = [110, 130, 150, 170, 190, 210]  
transition_spending = [4, 5.5, 6.5, 7.5, 8.5, 10]  
plt.figure(figsize=(10, 6))

plt.bar([year - 1 for year in years], world_gdp, width=1.5, label="PIL Mondo (Trillions USD)", color='#5e9cff')

plt.bar([year + 1 for year in years], transition_spending, width=1.5, label="Costo Annuale Transizione Ecologica (Trillions USD)", color='#5eff84')

plt.xlabel('Anno')
plt.ylabel('Trillions USD')
plt.title('PIL Mondo vs Costo Transizione Ecologica Annuale (2025-2050)')
plt.legend()

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.show()
```
