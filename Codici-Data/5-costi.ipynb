```python
import matplotlib.pyplot as plt

countries = ["Germany", "United Kingdom", "France", "Italy"]
gdp_values = [4.07, 3.07, 2.78, 2.01]  # in trilioni di USD

countries.append("Total Ecological Transition")
gdp_values.append(150)

plt.figure(figsize=(12, 6))
plt.bar(countries, gdp_values, color=['blue', 'blue', 'blue', 'blue', 'green'])

plt.xlabel('Country / Total Ecological Transition')
plt.ylabel('Amount (Trillions USD)')
plt.title('GDP of Major European Countries vs Total Ecological Transition Cost (2025-2050)')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.show()
```

```python
years = [2025, 2030, 2035, 2040, 2045, 2050]
world_gdp = [110, 130, 150, 170, 190, 210]  
transition_spending = [4, 5.5, 6.5, 7.5, 8.5, 10]  
plt.figure(figsize=(10, 6))

plt.bar([year - 1 for year in years], world_gdp, width=1.5, label="World GDP (Trillions USD)", color='blue')

plt.bar([year + 1 for year in years], transition_spending, width=1.5, label="Ecological Transition Spending (Trillions USD)", color='green')

plt.xlabel('Year')
plt.ylabel('Amount (Trillions USD)')
plt.title('World GDP vs Ecological Transition Spending (2025-2050)')
plt.legend()

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.show()
```
